```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pymc3 as pm 
import pandas as pd
import theano
import arviz as az
import os
import seaborn as sns
from scipy.stats import cauchy
np.random.seed(123)
wd = os.getcwd()
save_figs = False
```

```python
# data settings
n = 1
theta = 2
sigma_x = np.sqrt(5)
sigma_y = 1
```

```python
# generate data. 
x_tilde = np.random.normal(loc = 0, scale =sigma_x, size = (n,))
X = x_tilde + theta
Y = cauchy.rvs(loc = X, scale = sigma_y)
```

```python
# build CP model.
with pm.Model() as CP_model:
    # priors.
    theta = pm.Flat("theta")
    x = pm.Normal("x", mu = theta, sd = sigma_x, shape = (n,))
    # Data likelihood
    y_like = pm.Cauchy('y_like', alpha=x, beta=sigma_y, observed=Y)
```

```python
with CP_model:
    step1 = pm.Metropolis(vars = [x, theta])
    CP_trace = pm.sample(draws=5000, tune=1000, step = [step1], chains = 1)[1000:]
    # uncomment for NUST sampling. 
    #CP_trace = pm.sample(draws=5000, tune=1000, chains = 1)[1000:]
```

```python
cp_fig = az.plot_trace(CP_trace, var_names=("theta", "x"))
fig = plt.gcf() 
if save_figs:
    fig.savefig(wd+"/figures/cauchy_MH_cp_trace.pdf") 
```

```python
az.summary(CP_trace)
```

```python
# build NPC model.
with pm.Model() as NPC_model:
    # priors.
    theta = pm.Flat("theta")
    x_tilde = pm.Normal("x_tilde", mu = theta, sd = sigma_x, shape = (n,))
    x = pm.Deterministic("x", theta + x_tilde)
    # Data likelihood
    y_like = pm.Cauchy('y_like', alpha=x, beta=sigma_y, observed=Y)
```

```python
with NPC_model:
    step1 = pm.Metropolis(vars = [theta, x_tilde, x])
    NPC_trace = pm.sample(draws=5000, tune=1000, step = [step1], chains = 1)[1000:]
    # uncomment for NUST sampling.
    #NPC_trace = pm.sample(draws=5000, tune=1000, chains = 1)[1000:]
```

```python
npc_fig = az.plot_trace(NPC_trace, var_names=("theta", "x"))
fig = plt.gcf() # to get the current figure...
if save_figs:
    fig.savefig(wd+"/figures/cauchy_MH_npc_trace.pdf") # and save it directly
```
